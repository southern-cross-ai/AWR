```python
import requests

import PyPDF2
```

```python
prefix = 'https://www.womenaustralia.info/wp-admin/admin-ajax.php?action=efront_entry_pdf_export&id='
save_path_prefix = '/Users/yifan/Documents/GitHub/datasets/AWR/AWR'
```

```python
# last time checked: 22616 (17 Aug 2024)
max_id = 30000
skip_mask = r"b'\t<style>\n\t\t*,\n\t\t*::before,\n\t\t*::after"

# update the range od ID. e.g. range(30000, 50000)
for id in range(10348, 23001):
    url = prefix + str(id)
    response = requests.get(url)
    if response.status_code == 200:
        pdf_name = 'awr_' + str(id) + '.pdf'
        txt_name = 'awr_' + str(id) + '.txt'
        # skip blank files
        if str(response.content).startswith(skip_mask):
            print(f"Skipped blank {pdf_name}")
            continue
        # save as PDF files
        with open(save_path_prefix + '/' + pdf_name, "wb") as f:
            f.write(response.content)
            print(f"Saved as {pdf_name}")
        # # save as TXT files
        # with open(save_path_prefix + '/' + pdf_name, 'rb') as f:
        #     pdf = PyPDF2.PdfReader(f)
        #     text = ''
        #     for page in pdf.pages:
        #         text += page.extract_text()
        # with open(save_path_prefix + '/' + txt_name, 'w') as f:
        #     f.write(text)
    else:
        print(f"Failed to download {pdf_name}. Status code: {response.status_code}.")
```
